## BMI Analysis

```python
# =============================================================================
# 12. BMI ANALYSIS (WHO Guidelines)
# =============================================================================
print("\n" + "="*80)
print("STEP 12: BMI ANALYSIS (WHO Guidelines)")
print("="*80)

print("\nBMI Statistics:")
print("-" * 60)
print(df['BMI'].describe().round(2))

# Create BMI categories according to WHO guidelines
def categorize_bmi_who(bmi):
    """
    Categorize BMI according to WHO guidelines:
    - Underweight: BMI < 18.5
    - Normal weight: 18.5 â‰¤ BMI < 25
    - Pre-obesity: 25 â‰¤ BMI < 30
    - Obesity class I: 30 â‰¤ BMI < 35
    - Obesity class II: 35 â‰¤ BMI < 40
    - Obesity class III: BMI â‰¥ 40
    """
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal weight'
    elif bmi < 30:
        return 'Pre-obesity'
    elif bmi < 35:
        return 'Obesity class I'
    elif bmi < 40:
        return 'Obesity class II'
    else:
        return 'Obesity class III'

df['BMI_Category'] = df['BMI'].apply(categorize_bmi_who)

print("\nBMI Category Distribution (WHO Guidelines):")
print("-" * 60)
bmi_cat_counts = df['BMI_Category'].value_counts()
bmi_cat_order = ['Underweight', 'Normal weight', 'Pre-obesity', 
                 'Obesity class I', 'Obesity class II', 'Obesity class III']

for cat in bmi_cat_order:
    if cat in bmi_cat_counts.index:
        count = bmi_cat_counts[cat]
        pct = count / len(df) * 100
        print(f"  {cat:18s}: {count:6,} ({pct:5.2f}%)")

```

```python
# Visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# BMI distribution histogram
axes[0, 0].hist(df['BMI'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
axes[0, 0].axvline(df['BMI'].mean(), color='red', linestyle='--', linewidth=2, 
                   label=f'Mean: {df["BMI"].mean():.2f}')
axes[0, 0].axvline(df['BMI'].median(), color='orange', linestyle='--', linewidth=2, 
                   label=f'Median: {df["BMI"].median():.2f}')

# Add WHO threshold lines
axes[0, 0].axvline(18.5, color='green', linestyle=':', linewidth=1.5, alpha=0.7, label='WHO: 18.5')
axes[0, 0].axvline(25, color='yellow', linestyle=':', linewidth=1.5, alpha=0.7, label='WHO: 25')
axes[0, 0].axvline(30, color='orange', linestyle=':', linewidth=1.5, alpha=0.7, label='WHO: 30')
axes[0, 0].axvline(35, color='red', linestyle=':', linewidth=1.5, alpha=0.7, label='WHO: 35')
axes[0, 0].axvline(40, color='darkred', linestyle=':', linewidth=1.5, alpha=0.7, label='WHO: 40')

axes[0, 0].set_title('BMI Distribution with WHO Thresholds', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('BMI', fontsize=12)
axes[0, 0].set_ylabel('Frequency', fontsize=12)
axes[0, 0].legend(fontsize=8, loc='upper right')
axes[0, 0].grid(axis='y', alpha=0.3)

# BMI by diabetes status (box plot)
df.boxplot(column='BMI', by=target_variable, ax=axes[0, 1], patch_artist=True, color = {'boxes': 'lightblue', 'whiskers': 'black', 'medians': 'red', 'caps': 'black'})
axes[0, 1].set_title('BMI by Diabetes Status', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Diabetes Status', fontsize=12)
axes[0, 1].set_ylabel('BMI', fontsize=12)
axes[0, 1].set_xticklabels([label_mapping['Diabetes_012'][i]
                            for i in sorted(df[target_variable].unique())])
axes[0, 1].grid(axis='y', alpha=0.3)
plt.suptitle('')

# Ensure BMI_Category exists (some execution orders may skip the cell that creates it)
if 'bmi_cat_order' not in globals():
    bmi_cat_order = ['Underweight', 'Normal weight', 'Pre-obesity', 
                     'Obesity class I', 'Obesity class II', 'Obesity class III']

if 'BMI_Category' not in df.columns:
    # Create BMI categories according to WHO thresholds using pd.cut
    bins = [-float('inf'), 18.5, 25, 30, 35, 40, float('inf')]
    df['BMI_Category'] = pd.cut(df['BMI'], bins=bins, labels=bmi_cat_order, right=False)

# BMI category distribution (WHO)
bmi_cat_counts_ordered = df['BMI_Category'].value_counts().reindex(bmi_cat_order)
colors_who = ['lightblue', 'lightgreen', 'yellow', 'orange', 'red', 'darkred']
bmi_cat_counts_ordered.plot(kind='bar', ax=axes[1, 0],
                             color=colors_who,
                             edgecolor='black')
axes[1, 0].set_title('BMI Category Distribution (WHO Guidelines)', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('BMI Category', fontsize=12)
axes[1, 0].set_ylabel('Count', fontsize=12)
axes[1, 0].set_xticklabels(bmi_cat_order, rotation=45, ha='right')
axes[1, 0].grid(axis='y', alpha=0.3)

# Add count labels on bars
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt='%d', fontsize=9)

# BMI category by diabetes status (stacked bar)
cross_tab_bmi = pd.crosstab(df['BMI_Category'], df[target_variable], normalize='index') * 100
cross_tab_bmi = cross_tab_bmi.reindex(bmi_cat_order)
cross_tab_bmi.plot(kind='bar', stacked=True, ax=axes[1, 1],
                   color=['steelblue', 'orange', 'forestgreen'], edgecolor='black')
axes[1, 1].set_title('Diabetes Status by BMI Category (WHO)', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('BMI Category', fontsize=12)
axes[1, 1].set_ylabel('Percentage (%)', fontsize=12)
axes[1, 1].set_xticklabels(bmi_cat_order, rotation=45, ha='right')
axes[1, 1].legend([label_mapping['Diabetes_012'][i] for i in [0, 1, 2]],
                  title='Diabetes Status', fontsize=9)
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(f"{output_dir}/06_bmi_analysis.png", dpi=300, bbox_inches='tight')
plt.show()
print(f"\nSaved: {output_dir}/06_bmi_analysis.png")

# Additional analysis: Mean BMI by diabetes status
print("\nMean BMI by Diabetes Status:")
print("-" * 60)
for cls in sorted(df[target_variable].unique()):
    mean_bmi = df[df[target_variable] == cls]['BMI'].mean()
    print(f"  {label_mapping['Diabetes_012'][cls]:12s}: {mean_bmi:.2f}")

# Drop temporary BMI_Category column if we created it in this cell
try:
    df.drop('BMI_Category', axis=1, inplace=True)
except KeyError:
    pass
```
